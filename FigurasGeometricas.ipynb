```python
class FiguraGeometrica:
    def __init__(self, ancho, alto):
        self._ancho = None
        self._alto = None
        self.ancho = ancho
        self.alto = alto

    #encapsulamiento de ancho
    @property
    def ancho(self):
        return self._ancho

    @ancho.setter
    def ancho(self, valor):
        if valor <= 0:
            raise ValueError("El ancho debe ser mayor que 0")
        self._ancho = valor

    #encapsulamiento de alto
    @property
    def alto(self):
        return self._alto

    @alto.setter
    def alto(self, valor):
        if valor <= 0:
            raise ValueError("El alto debe ser mayor que 0")
        self._alto = valor

    def area(self):
        return self._ancho * self._alto

    def perimetro(self):
      pass

    def __str__(self):
        return f"Ancho: {self.ancho}, Alto: {self.alto}"
```

```python
#clase cuadrado
class Cuadrado(FiguraGeometrica):
    def __init__(self, lado):
        FiguraGeometrica.__init__(self, lado, lado)
#encapsulamiento de lado
    @property
    def lado(self):
        return self.ancho

    @lado.setter
    def lado(self, valor):
        self.ancho = valor
        self.alto = valor

    def area(self):
        return self.lado ** 2

    def perimetro(self):
        return 4 * self.lado

    def __str__(self):
        return f"Cuadrado -> Lado: {self.lado}"
```

```python
#clase rectangulo
class Rectangulo(FiguraGeometrica):
    def __init__(self, ancho, alto):
        FiguraGeometrica.__init__(self, ancho, alto)

    def area(self):
        return self.ancho * self.alto

    def perimetro(self):
        return 2 * (self.ancho + self.alto)

    def __str__(self):
        return f"Rectángulo -> Ancho: {self.ancho}, Alto: {self.alto}"
```

```python
#polimorfismo
def sumar_areas(figuras):
    total = 0
    for figura in figuras:
        total += figura.area()
    return total


def sumar_perimetros(figuras):
    total = 0
    for figura in figuras:
        total += figura.perimetro()
    return total
```

```python
# programa principal

# Crear objetos
c1 = Cuadrado(4)
c2 = Cuadrado(2)

r1 = Rectangulo(3, 5)
r2 = Rectangulo(6, 2)

# Lista de figuras
figuras = [c1, r1, c2, r2]

# Mostrar datos
for figura in figuras:
    print(figura)
    print("Área:", figura.area())
    print("Perímetro:", figura.perimetro())
    print()

# Modificar valores (encapsulamiento)
c1.lado = 10
print("Cuadrado modificado:", c1)

r1.ancho = 8
r1.alto = 4
print("Rectángulo modificado:", r1)

print()

# Validación de error
try:
    c2.lado = -3
except ValueError as e:
    print("Error detectado:", e)

print()

# Polimorfismo
print("Suma total de áreas:", sumar_areas(figuras))
print("Suma total de perímetros:", sumar_perimetros(figuras))
```
